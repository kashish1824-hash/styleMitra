```python
selected_df = df[
    [
        "Hair Color",
        "Eye Color",
        "Skin Tone",
        "Under Tone",
        "Torso length",
        "Body Proportion",
        "Recommended Clothing Colors",
        "Avoid Clothing Colors",
        "Recommended Fitting Style",
        "Recommended Materials",
        "Recommended Patterns",
        "Recommended Jewelry Metal",
        "Recommended Shoes",
        "Recommended Clothing Color Wheel Region"
    ]
]

selected_df.to_csv(
    "../data/stylist_dataset.csv",
    index=False
)

print(selected_df.shape)
```

```python
X = selected_df[
    [
        "Hair Color",
        "Eye Color",
        "Skin Tone",
        "Under Tone",
        "Torso length",
        "Body Proportion"
    ]
]

y = selected_df[
    [
        "Recommended Clothing Colors",
        "Avoid Clothing Colors",
        "Recommended Fitting Style",
        "Recommended Materials",
        "Recommended Patterns",
        "Recommended Jewelry Metal",
        "Recommended Shoes",
        "Recommended Clothing Color Wheel Region"
    ]
]

print(X.head())

print(y.head())
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}

X_encoded = X.copy()

for column in X.columns:

    le = LabelEncoder()

    X_encoded[column] = le.fit_transform(
        X[column]
    )

    encoders[column] = le

X_encoded.head()
```

```python
y_encoded = y.copy()

target_encoders = {}

for column in y.columns:

    le = LabelEncoder()

    y_encoded[column] = le.fit_transform(
        y[column]
    )

    target_encoders[column] = le

y_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)

print(X_test.shape)

print(y_train.shape)

print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier

model = MultiOutputClassifier(
    RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )
)

model.fit(
    X_train,
    y_train
)

print("Training complete")
```

```python
sample = X_test.iloc[[0]]

prediction = model.predict(sample)

prediction
```

```python
decoded_output = {}

for i, column in enumerate(y.columns):

    decoded_output[column] = target_encoders[
        column
    ].inverse_transform(
        [prediction[0][i]]
    )[0]

decoded_output
```

```python
from sklearn.metrics import accuracy_score

predictions = model.predict(
    X_test
)

for i, column in enumerate(y.columns):

    acc = accuracy_score(
        y_test.iloc[:, i],
        predictions[:, i]
    )

    print(
        column,
        ":",
        round(acc, 3)
    )
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(
    X_encoded
)

kmeans = KMeans(
    n_clusters=5,
    random_state=42
)

clusters = kmeans.fit_predict(
    X_scaled
)

selected_df["Cluster"] = clusters

selected_df[
    [
        "Hair Color",
        "Skin Tone",
        "Under Tone",
        "Cluster"
    ]
].head()
```

```python
selected_df = selected_df.copy()

selected_df["Cluster"] = clusters
```

```python
selected_df.groupby(
    "Cluster"
)[
    [
        "Skin Tone",
        "Under Tone",
        "Recommended Jewelry Metal"
    ]
].agg(
    lambda x: x.mode()[0]
)
```

```python
def get_cluster_recommendation(
    cluster_id
):

    return selected_df[
        selected_df[
            "Cluster"
        ] == cluster_id
    ][
        [
            "Recommended Clothing Colors",
            "Recommended Jewelry Metal",
            "Recommended Materials",
            "Recommended Shoes"
        ]
    ].mode().iloc[0]


get_cluster_recommendation(
    3
)
```
